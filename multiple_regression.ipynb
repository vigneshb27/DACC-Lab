```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
df = pd.read_csv('Life_Expectancy_Data.csv')
df.head()
```

```python
df = df.dropna(subset = ['Alcohol', 'Life_expectancy', 'Adult_Mortality', 'BMI', 'Total_expenditure', 'under_five_deaths'])
```

```python
df.isnull().sum()
```

```python
df = df.drop('Country', axis=1)
df = df.drop('Status', axis=1)
```

```python
sns.heatmap(df.corr())
```

```python

independent_vars = ['percentage_expenditure', 'Adult_Mortality', 'infant_deaths', 'BMI', 'under_five_deaths', 'Alcohol', 'Measles']
features = df[independent_vars]
res = df['Life_expectancy']
X_train,X_test,y_train,y_test=train_test_split(features,res,test_size=0.3)
```

```python
model=LinearRegression()
fittedModel=model.fit(X_train,y_train)
predictions=fittedModel.predict(X_test)
fittedModel.score(X_test,y_test)
```
